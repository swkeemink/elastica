Basic model parameters

```python
Ac = 1     # basic tuning magnitude
Kc = 1   # basic tuning width (lower is wider)
a  = 0.1  # modulation strength
E0 = 4   # modulation offset
N  = 32    # number of neurons per location
```

Plotting function

```python
def plot_scenes(dist, A, X, saltype):
    # Setup the scene
    scene_sur = en.scene(N,A,X,[X[:,0].max()+dist,X[:,1].max()+dist],Kc,Ac,a,E0)

    # simulate fixed, and cdep, to get max saliency
    modtype='fixed'
    scene_sur.simulate_all(torus='on',modtype=modtype)
    scene_sur.saliency(saltype)
    maxsal = scene_sur.sal.max()
    print maxsal
    modtype='cdep'
    scene_sur.simulate_all(torus='on',modtype=modtype)
    scene_sur.saliency(saltype)
    maxsal = max(maxsal,scene_sur.sal.max())
    print maxsal
    
    # simulate fixed
    modtype='fixed'
    scene_sur.simulate_all(torus='on',modtype=modtype)

    # calculate saliency
    scene_sur.saliency(saltype)
    # initiate figure
    fig = scene_sur.plotscene(length=dist/2) 
    fig+= scene_sur.plotscene(length=dist/2,oriens=scene_sur.est,alphas=scene_sur.sal/maxsal)
    # simulate center dependent
    modtype='cdep'
    scene_sur.simulate_all(torus='on',modtype=modtype)
#     fig+= hv.Curve(scene_sur.sal)
    # calculate saliency
    scene_sur.saliency(saltype)
#     fig+= hv.Curve(scene_sur.sal)
    # add to figure
    fig+= scene_sur.plotscene(length=dist/2,oriens=scene_sur.est,alphas=scene_sur.sal/maxsal)

    return fig
```

Single popout

```python
%%opts Curve {+axiswise} (linewidth=8) Overlay [xticks=0 yticks=0]
%%output fig='pdf'
# set some scene parameters
ns = 7 # width of height of scene in bars
dist = 5 # distance between bars
a = 0.25
# specify saliency type
saltype = 'mean'

# set bar orientations
feat = 45*pi/180 # feature bar orientation
back = 0*pi/180 # background bars orientation
A  = np.ones(ns*ns)*back # define all bars as having backgound degree orientation
feat_ids = ns*ns/2 # choose feature
A[feat_ids] = feat

# set bar positions (x and y positions)
X = np.array([[x,y] for x in np.arange(ns)*dist for y in np.arange(ns)*dist]) # set all bars positions

fig_single = plot_scenes(dist, A, X, saltype)
fig_single
```

Contour

```python
%%opts Curve {-axiswise} (linewidth=8 color='k') Overlay [xticks=0 yticks=0]
# set some scene parameters
ns = 7 # width of height of scene in bars
dist = 5 # distance between bars
a=0.25
# specify saliency type
saltype = 'max'

# set bar orientations
feat = 45*pi/180 # feature bar orientation
back = 0*pi/180 # background bars orientation
A  = np.ones(ns*ns)*back # define all bars as having backgound degree orientation
feat_ids = [ns*(i)+(i) for i in range(ns)] # choose feature
A[feat_ids] = feat

# set bar positions (x and y positions)
X = np.array([[x,y] for x in np.arange(ns)*dist for y in np.arange(ns)*dist]) # set all bars positions

fig_cont = plot_scenes(dist,A, X, saltype)
fig_cont
```

Salient feature

```python
%%opts Curve {+axiswise} (linewidth=8) Overlay [xticks=0 yticks=0]
# set some scene parameters
ns = 9 # width of height of scene in bars
dist = 5 # distance between bars

# specify saliency type
saltype = 'max'

# set bar orientations
feat = 90*pi/180 # feature bar orientation
back = 0*pi/180 # background bars orientation
A  = np.ones(ns*ns)*back # define all bars as having backgound degree orientation
# choose feature
feat_ids = []
for i in range(3):
    feat_ids += [3+(3+i)*ns,4+(3+i)*ns,5+(3+i)*ns] 
A[feat_ids] = feat

# set bar positions (x and y positions)
X = np.array([[x,y] for x in np.arange(ns)*dist for y in np.arange(ns)*dist]) # set all bars positions

fig_feat = plot_scenes(dist,A, X, saltype)
fig_feat
```

Edge

```python
%%opts Curve {+axiswise} (linewidth=4) Overlay [xticks=0 yticks=0]
# set some scene parameters
ns_x = 20 # width of scene in bars
ns_y = ns_x/2 # height
dist = 3 # distance between bars

# specify saliency type
saltype = 'mean'

# set bar orientations
feat = 45*pi/180 # feature bar orientation
back = 90*pi/180 # background bars orientation
A  = np.ones(ns_x*ns_y)*back # define all bars as having backgound degree orientation
feat_ids = range(ns_x*ns_y/2) # choose feature
A[feat_ids] = feat

# set bar positions (x and y positions)
X = np.array([[x,y] for x in np.arange(ns_x)*dist for y in np.arange(ns_y)*dist]) # set all bars positions

fig_edge = plot_scenes(dist, A, X, saltype)
fig_edge
```
